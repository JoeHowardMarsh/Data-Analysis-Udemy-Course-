```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()

# Pandas library is used for importing various file types e.g. csv, excel, json.
# also used for manipulation and cleaning dataframes.

# Numpy library has functions used for working with arrays and matrices
# faster, clearer and better quality code using Numpy

# Matplotlib is a 2D plotting library for visualising data in Python

# Seaborn is used for improving Matplotlib visualisations to make them more visually attractive.
# It builds on top of MatPlotLib and works with Pandas for data structures

# sklearn is a library used for machine learning and statistical modeling.
# it includes many tools for classification, regression, clustering and dimensionality reduction
```

```python
raw_data = pd.read_csv("LR Real Life Example.csv")
raw_data.head()
# importing and displaying the csv file data
```

### Pre-processing

```python
raw_data.describe(include='all')
# looking at descriptive statistics
```

```python
data = raw_data.drop(['Model'],axis=1)
data.describe(include='all')

# as the 'Registration' entry has 3947 yes entries, this variable won't be useful
# implementing 'Model' variable into the regression will be very tedious so the variable can be dropped
# most of this info can come from other variables
```

```python
# here we can see that Brand, body, mileage, engine type, registration and year all have 4345 values,
# ... but 'price' and 'engineV' have some missing values
```

### Dealing with Missing Values

```python
data.isnull().sum() 
# this shows the dataframe with 'null' values
# the .sum() at the end converts False into 0 and True into 1, now we know that 'price' has 172 missing values 
# and 'EngineV' has 150 missing values
```

```python
# if less than 5% of all values have missing observations, it's safe to remove them

data_no_mv = data.dropna(axis=0)

# new object created to get data with no missing values
# axis is 0 as we are dropping observations (rows), not columns (variables)
```

```python
data_no_mv.describe(include='all')

# now the count of each feature is the same (4025) so there are no longer any missing values
```

### Exploring the PDFs (Probability Distribution Functions)

```python
sns.displot(data_no_mv['Price'])

# using seaborn module to plot price distribution
```

### Dealing with Outliers

```python
# there are many outliers towards the upper end of the distribution
# 75% of all entries are below 21,900, but the prices range up to 300,000
# this will cause the coefficients to be inflated and therefore should be removed
# a good way to do this is by removing the top 1% of observations using the Quantile Method

q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')

# here we are getting the 99 percentile and keeping data below the 99 percentile
# with only the top 1% removed, our distribution range is almost 3 times smaller
```

```python
sns.displot(data_1['Price'])

# plot of the price distribution again, this time without the outliers
```

```python
sns.displot(data_no_mv['Mileage'])

# we can see the same effect from outliers is occurring in Mileage
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]

# using the same method to remove the top 1% of observations to deal with outliers
```

```python
sns.displot(data_2['Mileage'])

# plot of the mileage distribution range after removal of outliers
```

```python
EngV = pd.DataFrame(raw_data['EngineV'])
EngV = EngV.dropna(axis=0)
EngV.sort_values(by='EngineV')

# as no cars have engine volumes of 99, the maximum is around 6.5 litres, we know these entries must be wrong
```

```python
data_3 = data_2[data_2['EngineV']<6.5]
sns.displot(data_3['EngineV'])

# as the range of engine volumes lies below 6.5, it is safe to remove any entries with volumes above that threshold
```

```python
sns.displot(data_no_mv['Year'])

# displaying the 'year' distribution of observations
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]

# this time, the outliers are on the low end of the 'Year' axis (due to vintage cars)
# so we want to take the first percentile and keep any observations that are higher
```

```python
sns.displot(data_4['Year'])

# displaying the distribution of 'year' after removing the outliers
```

```python
data_cleaned = data_4.reset_index(drop=True)

# now that the data is cleaned, a new variable is created to store the new data and reset the indexes
```

```python
data_cleaned.describe(include='all')
```

### Multicollinearity

```python
# definition = one predictor variable in a multiple regression model can be linearly predicted
# from the other variables with a substantial degree of accuracy
```

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns

# using the VIF (Variance Inflation Factor) to check for multicollinearity
# creating an object with the variables to check for multicollinearity (mileage, year and engineV)
```

```python
vif

# if VIF = 1, there is no multicollinearity
# generally VIF values between 1 and 5 are good
# it's debatable whether VIF value over 5 is ok, but over VIF 10 is not ok
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)

# therefore we should drop 'Year' as it's VIF is over 10
```

### Creating Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)

# to add categorical data to a regression, they must be converted into continuous dummy variables
```

```python
data_with_dummies.head()
```

### Rearranging

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']

# rearranging the features so that 'log_price' appears first
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear Regression Model

### Declare the inputs and the targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)

# storing the log_price in the variable 'targets'
# storing the rest of the features except 'log_price' in the variable inputs
```

### Scale the Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)

# using a module from sklearn called 'StandardScaler' to scale the inputs
```

```python
inputs_scaled = scaler.transform(inputs)

# storing the scaled inputs into a new variable called 'inputs_scaled'
```

### Train Test Split 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)

# splitting our input and target data into training and testing (80% training, 20% testing)
```

### Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)

# storing the linear regression class into the variable reg, then fitting the data into that class
```

```python
y_hat = reg.predict(x_train)

# the predictions of the linear model are labelled with y_hat
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()

# plotting the predictions (y_hat) against the targets (y_train)
```

```python
sns.displot(y_train - y_hat)
plt.title("Residuals PDF", size=18)

# displaying the difference between the targets and the predictions (residuals)
# the results show a relatively normal distribution, however there is a much longer tail on the negative side
```

```python
reg.score(x_train, y_train)

# calculating the R-Squared
# (the proportion of the variance for a dependent variable that's explained by the independent variable)
# the model explains around 75% of the variability of the data
```

### Finding Weights and Bias

```python
reg.intercept_

# the intercept is the point where the function crosses the y axis (the predicted value for y when x is 0)
```

```python
reg.coef_

# this is the constant that is multiplied by the variable for each feature
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features']) 
reg_summary['Weights'] = reg.coef_
reg_summary

# displaying the weights
```

Weights interpretation
1. a positive weight shows that as a feature increases in value, so does the log_price and 'Price' respectively
2. a negative weight shows that as a feature decreases in value, so does the log_price and 'Price' respectively

Dummy Variables

2. a negative weight shows that the respective category (Brand) is less expensive than the benchmark (Audi)

### Testing

```python
y_hat_test = reg.predict(x_test)

# storing the predictions of the model in the y_hat test variable
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.3)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()

# plotting the results of the predictions against the targets
# our model is better at predicting when it comes to higher prices, however for lower prices the results are scattered
# alpha turns the graph into a heatmap showing the concentration of values in a given area
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()

# creating a new dataframe measuring performance of the regression model
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()

# there are a lot of missing values 
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()

# resetting the index of y_test
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()

# now we have those missing values
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']

# comparing the predictions and the targets
# the residual will be the difference between the targets and the predictions
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf

# this retrieves the percentage difference between the predictions and the targets
```

```python
df_pf.describe()

# showing descriptive stats of the dataframe
# minimum diff. in percentages is 0.06%, however the maximum difference is over 500% 
```
